```python
import pandas as pd
import numpy as np

data = pd.read_csv("classification.csv")
data.head()
```

```python
data.describe()
```

#### Requires you to perform scaling, with respect to the mean and std or min and max scalars

### Rule of thumb : Distance based or Log based regression would require scaling of the data

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

### Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train1 = sc.fit_transform(x_train)
x_test1 = sc.transform(x_test)
x_train1
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(x_train1,y_train)
```

```python
y_pred1 = log_reg.predict(x_test1)
y_pred1
```

### Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()

x_train2 = mm.fit_transform(x_train)
x_test2 = mm.transform(x_test)
x_train2
```

```python
log_reg2 = LogisticRegression()
log_reg2.fit(x_train2,y_train)
y_pred2 = log_reg2.predict(x_test2)
y_pred2
```

#### Accuracy for Std Scalar 

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test,y_pred1)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred1))
```

#### Accuracy for MinMax Scalar

```python
confusion_matrix(y_test,y_pred2)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
print(classification_report(y_test,y_pred2))
```
